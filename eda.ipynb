```python
import os
import pandas as pd
from configs.config import cfg

eda_dir = cfg["eda_dir"]    
```

```python
df = pd.read_csv(cfg['data']['raw_root']+ "/Sampled_Sarcopenia_Data.csv")

df.head()
```

```python
# rename F -> Female and M -> Male
```

```python
gender_counts = df["Gender"].value_counts()

def append_row_counts(output_data, row_counts):
    for row, count in row_counts.items():
        output_data.append([row, count])
    return output_data

output_data = [
    ["Sex", ""],
]

output_data = append_row_counts(output_data, gender_counts)
output_df = pd.DataFrame(output_data, columns=["Variable", "Count"])
output_df.to_excel(os.path.join(eda_dir,'Demographics.xlsx'), index=False)
```
